ETL - Extract, Transform, Load

Extract: Our data is extracted from an Excel file named 'Additional Hospitalization deta.xlsx' . In the ETL process, the data is first extracted so that we can work on it.

```python
Additional_Hospitalization_df = pd.read_excel('Additional Hospitalization deta.xlsx')
Additional_Hospitalization_df
```

Transform: The ETL's central process is in which various changes are made to the original data so that we can adapt the data to the project's goals. In addition, in the process, we clean the data of empty values ​​and noise so that they do not interfere with drawing conclusions from the data.

```python
Additional_Hospitalization_df.isnull().sum()
```

```python
Additional_Hospitalization_df = Additional_Hospitalization_df.dropna()
Additional_Hospitalization_df
```

```python
Additional_Hospitalization_df.duplicated().sum()
```

```python
Additional_Hospitalization_df.drop_duplicates(inplace=True)
Additional_Hospitalization_df
```

Sort the DF by Patient Number and then by date

```python
Additional_Hospitalization_df.sort_values(by=['Patient Number', 'Date'], inplace=True)
Additional_Hospitalization_df
```

Load: Loading all the tables and merging them into one final table. Because in our project we worked only on a Fact table, we will present the final table after the Transform.
